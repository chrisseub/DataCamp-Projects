```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
# Start coding here! Use as many cells as you like
#Viewing the Datafame
netflix_df.describe()
netflix_df.shape
```

```python
# Filter the data for movies released in the 1990s
movies_90s = netflix_df[(netflix_df['type'] == 'Movie') & (netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] < 2000)]
movies_90s
```

```python
# Visualizing the distribution of durations for movies released in the 1990s
plt.hist(movies_90s['duration'], bins=20, edgecolor='black')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.title('Distribution of Durations for 1990s Movies')
plt.show()
```

```python
# Subset the data to keep only movies with a specific genre
comedy_movies_90s = movies_90s[movies_90s['genre'] == 'Comedy']
action_movies_90s = movies_90s[movies_90s['genre'] == 'Action']
drama_movies_90s = movies_90s[movies_90s['genre'] == 'Dramas']
doc_movies_90s = movies_90s[movies_90s['genre'] == 'Documentaries']

```

```python
# Count the number of movies with a duration less than 90 minutes 

short_comedy_movies_count = len(comedy_movies_90s[comedy_movies_90s['duration'] < 90])  
short_comedy_movies_count 
```

```python
# Filter the data for action movies released in the 1990s with duration less than 90 minutes
short_action_movies_90s = movies_90s[(movies_90s['type'] == 'Movie') & 
                                     (movies_90s['release_year'] >= 1990) & 
                                     (movies_90s['release_year'] < 2000) & 
                                     (movies_90s['genre'] == 'Action') & 
                                     (movies_90s['duration'] < 90)]

short_action_movies_count = len(short_action_movies_90s)
short_action_movies_count
```

```python
# Initialize the counter for short action movies in the 1990s
short_movie_count = 0

# Iterate over the movies in the 1990s
for index, row in movies_90s.iterrows():
    if (row['release_year'] >= 1990 and row['release_year'] < 2000 and 
        row['type'] == 'Movie' and row['genre'] == 'Action' and 
        row['duration'] < 90):
        short_movie_count += 1

short_movie_count
```
